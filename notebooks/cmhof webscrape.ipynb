```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import re
```

```python
page = 1
artists = []
#years = []
while page != 18:
      url = f"https://countrymusichalloffame.org/hall-of-fame/members/page/{page}"
      response = requests.get(url)
      html = response.content
      soup = BS(html, "lxml")
      for h3 in soup.find_all("h3"):
        artists.append(h3.get_text(strip=True))
      page = page + 1
      
```

```python
artists
```

```python
page = 1
year = []
while page != 18:
      url = f"https://countrymusichalloffame.org/hall-of-fame/members/page/{page}"
      response = requests.get(url)
      html = response.content
      soup = BS(html, "lxml")
      years1 = str(soup.findAll(attrs={'class' : 'vertical-card_content--copy'}))
      years1 = re.findall(r"\d{4}", years1)
      year.append(years1)
      page = page + 1
```

```python
print(type(year))
flat_list=list()
for sub_list in year:
    flat_list += sub_list
year = flat_list
year
```

```python
df1 = {'Artist':artists,'Year':year}
df1 = pd.DataFrame(df1)
df1
```
